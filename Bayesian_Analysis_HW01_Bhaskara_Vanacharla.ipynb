```python
# Define the initial probability distribution
prior_probabilities = [0.0, 0.25, 0.5, 0.75, 1.0]

# Observed pattern
observed_pattern = ['blue', 'white', 'blue']

# Define the likelihood function for the observed pattern
def likelihood(observed_pattern, probability_blue):
    plausibility = 1.0
    for ball in observed_pattern:
        if ball == 'blue':
            plausibility *= probability_blue
        else:
            plausibility *= (1 - probability_blue)
    return plausibility

# Define the prior probability function
def prior_probability(probability_blue):
    return 1.0  # Uniform prior for simplicity

# Define the posterior probability function using Bayes' Theorem
def posterior_probability(observed_pattern, probability_blue):
    prior = prior_probability(probability_blue)
    likelihood_value = likelihood(observed_pattern, probability_blue)
    evidence = sum(likelihood(observed_pattern, p) * prior_probability(p) for p in prior_probabilities)
    posterior = (likelihood_value * prior) / evidence
    return posterior

# Calculate and print posterior probabilities for each prior probability
for prior_probability_blue in prior_probabilities:
    posterior = posterior_probability(observed_pattern, prior_probability_blue)
    print(f"Prior probability of blue ball: {prior_probability_blue}")
    print(f"Posterior probability for observing the pattern: {posterior}")
    print("---------------------")
```
